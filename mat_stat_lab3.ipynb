# Лабораторная работа №3
 Проверка гипотез о параметрах нормального распределения





```python
import pandas as pd
import statistics
import numpy as np
import pandas
import scipy.stats as stats
import statistics as stat
import statsmodels

df = pd.DataFrame(pd.read_excel("/content/mat_stat_3(14)1.xlsx"))
df
```

```python
var_x1 = 36
#var_x2 = undefined
a_0 = 40
sd_0 = 4
var_y1 = 9
#var_y2 = und
alpha = 0.05
```

## Пункт №1

Гипотеза о значении математического ожидания при известной дисперсии (one-sampled z-test).

```python
n = len(df['X1'])
z_score = (stat.fmean(df["X1"])-a_0)/(np.sqrt(var_x1)/np.sqrt(n))
z_critical = stats.norm.ppf(1-alpha)

if z_score >  z_critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")
```

## Пункт №2

Гипотеза о значении математического ожидания при неизвестной дисперсии (one-sample t-test)

```python
var = stat.variance(df["X2"], xbar=None)
n = len(df['X2'])
t_score = (stat.fmean(df["X2"])-a_0)/(np.sqrt(var)/np.sqrt(n))
t_critical = stats.t.ppf(1 - alpha, n - 1)

if t_score >  t_critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")
```

## Пункт №3

Гипотеза о значении дисперсии при неизвестном математическом ожидании (chi-squared test)

```python
var = stat.variance(df["X2"], xbar=None)
n = len(df['X2'])
statistic = (n - 1)*(stat.stdev(df['X2'])**2)/(sd_0**2)
critical = stats.chi2.ppf(1 - alpha, n - 1)

if statistic >  critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")
```

## Пункт №4

Гипотеза о равенстве математических ожиданий при известных дисперсиях (two-sample z-test)

```python
n1 = len(df['X1'])
n2 = len(df['Y1'])
statistic = (stat.fmean(df["X1"])-stat.fmean(df["Y1"]))/np.sqrt(var_x1/n1 + var_y1/n2)
critical = stats.norm.ppf(1 - alpha)

if statistic >  critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")
```

## Пункт №5

Гипотеза о равенстве математических ожиданий при неизвестных дисперсиях (two-sample unpooled t-test).

```python
n1 = len(df['X2'])
n2 = len(df['Y2'])
var = ((n1 - 1)*stat.stdev(df['X2'])**2 + (n2 - 1)* stat.stdev(df['Y2'])**2)/(n1 + n2 - 2)
sd = np.sqrt(var)

statistic = (stat.fmean(df["X2"])-stat.fmean(df["Y2"]))/(sd*np.sqrt(1/n1 + 1/n2))
critical = stats.t.ppf(1 - alpha, n1 + n2 - 2)

if statistic >  critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")
```

## Пункт №6


Гипотеза о равенстве дисперсий при неизвестных математических ожиданиях (two-sample F-test).

```python
n1 = len(df['X2'])
n2 = len(df['Y2'])

statistic = ((stat.stdev(df['X2'])**2)/(sd_0**2))/((stat.stdev(df['Y2'])**2)/(sd_0**2))
critical = stats.f.ppf(1 - alpha, n1 - 1, n2 - 1)

if statistic >  critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")
```

## Пункт №7

Гипотеза о равенстве математических ожиданий при неизвестных дисперсиях (two-sample unpooled t-test).

```python
n1 = len(df['Z1']) - df['Z1'].isnull().sum() #40
n2 = len(df['Z2']) - df['Z2'].isnull().sum()
n3 = len(df['Z3']) - df['Z3'].isnull().sum()
n4 = len(df['Z4']) - df['Z4'].isnull().sum() #45


var = ((n1 - 1)*stat.stdev(df['Z1'].dropna())**2 + (n2 - 1)* stat.stdev(df['Z2'])**2)/(n1 + n2 - 2)
sd = np.sqrt(var)
statistic = (stat.fmean(df['Z1'].dropna())-stat.fmean(df["Z2"]))/(sd*np.sqrt(1/n1 + 1/n2))
critical = stats.t.ppf(1 - alpha, n1 + n2 - 2)
if statistic >  critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")


var = ((n2 - 1)*stat.stdev(df['Z2'])**2 + (n3 - 1)* stat.stdev(df['Z3'])**2)/(n3 + n2 - 2)
sd = np.sqrt(var)
statistic = (stat.fmean(df["Z2"])-stat.fmean(df["Z3"]))/(sd*np.sqrt(1/n2 + 1/n3))
critical = stats.t.ppf(1 - alpha, n2 + n3 - 2)
if statistic >  critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")


var = ((n3 - 1)*stat.stdev(df['Z3'])**2 + (n4 - 1)* stat.stdev(df['Z4'].dropna())**2)/(n3 + n4 - 2)
sd = np.sqrt(var)
statistic = (stat.fmean(df["Z3"])-stat.fmean(df["Z4"].dropna()))/(sd*np.sqrt(1/n3 + 1/n4))
critical = stats.t.ppf(1 - alpha, n3 + n4 - 2)
if statistic >  critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")


var = ((n1 - 1)*stat.stdev(df['Z1'].dropna())**2 + (n4 - 1)* stat.stdev(df['Z4'].dropna())**2)/(n1 + n4 - 2)
sd = np.sqrt(var)
statistic = (stat.fmean(df["Z1"].dropna())-stat.fmean(df["Z4"].dropna()))/(sd*np.sqrt(1/n1 + 1/n4))
critical = stats.t.ppf(1 - alpha, n1 + n4 - 2)
if statistic >  critical:
    print("reject null hypothesis")
else:
  print("null hypothesis")
```
